## AI HUB 데이터 로드

```python
import json
import pandas as pd
import os
file_dir = '/content/drive/MyDrive/project3/한국어 방언 발화(제주도)/Training/[라벨]제주도_학습용데이터_1.zip (Unzipped Files)/'
filelist = os.listdir('/content/drive/MyDrive/project3/한국어 방언 발화(제주도)/Training/[라벨]제주도_학습용데이터_1.zip (Unzipped Files)')
standard_form = []
dialect_form=[]

# AI HUB 데이터 폴더 안에 json과 txt 파일이 섞여 있으므로, 예외처리를 통해 데이터 로드
for file in filelist:
    with open(file_dir+file,'r') as f:
        try:
            data = json.load(f)
            standard_form += [data['utterance'][i]['standard_form'] for i in range(len(data['utterance']))]
            dialect_form += [data['utterance'][i]['dialect_form'] for i in range(len(data['utterance']))]
        except json.JSONDecodeError:
            continue
```

```python
jejudata = pd.DataFrame()
```

```python
jejudata['standard'] = pd.DataFrame(standard_form)
jejudata['dialect'] = pd.DataFrame(dialect_form)
```

```python
jejudata.to_csv('/content/drive/MyDrive/project3/한국어 방언 발화(제주도)/Training/jeju_train.csv')
```

```python
v_file_dir = '/content/drive/MyDrive/project3/한국어 방언 발화(제주도)/Validation/[라벨]제주도_학습용데이터_3.zip (Unzipped Files)/'
v_filelist = os.listdir('/content/drive/MyDrive/project3/한국어 방언 발화(제주도)/Validation/[라벨]제주도_학습용데이터_3.zip (Unzipped Files)')
v_standard_form = []
v_dialect_form=[]

# AI HUB 데이터 폴더 안에 json과 txt 파일이 섞여 있으므로, 예외처리를 통해 데이터 로드
for file in v_filelist:
    with open(v_file_dir + file, 'r') as f:
        try:
            data = json.load(f)
            v_standard_form += [data['utterance'][i]['standard_form'] for i in range(len(data['utterance']))]
            v_dialect_form += [data['utterance'][i]['dialect_form'] for i in range(len(data['utterance']))]
        except json.JSONDecodeError:
            continue
```
